```python
import skimage
import skimage.io
import matplotlib.pyplot
%matplotlib inline
```

```python
kitten = skimage.io.imread('data/kittens.jpg')
```

```python
matplotlib.pyplot.imshow(kitten)
```

```python
type(kitten)
```

```python
kitten.shape
```

```python
blue = kitten[:,:,2]
```

```python
matplotlib.pyplot.imshow(blue)
```

```python
matplotlib.pyplot.imshow(blue, cmap=matplotlib.pyplot.cm.gray)
```

```python
import skimage.filter
import skimage.morphology
```

```python
kittens_blurred = skimage.filter.rank.median(blue,skimage.morphology.disk(15))
```

```python
matplotlib.pyplot.imshow(kittens_blurred, cmap=matplotlib.pyplot.cm.gray)
```

```python
kitten_mask = kittens_blurred > 80
```

```python
matplotlib.pylab.imshow(kitten_mask, cmap=matplotlib.pyplot.cm.gray)
```

```python
copy = kitten.copy()
```

```python
copy[kitten_mask == False] = (0,0,0)
```

```python
matplotlib.pyplot.imshow(copy, cmap=matplotlib.pyplot.cm.gray)
#matplotlib.image.AxesImage(copy)
```
