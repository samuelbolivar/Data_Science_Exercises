```python
import pandas
crime = pandas.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/communities/communities.data', 
                    header=None, na_values=['?'],
                        names=column_names,
                    index_col='communityname')
crime
```

```python
# The first interesting column is the 'population' column, and we are aiming to 
# predict the final column (ViolentCrimesPerPop)
# Create a suitable subset of the data frame to act as your X and y
y = crime.ViolentCrimesPerPop
X_columns = [x for x in crime.columns if x != 'ViolentCrimesPerPop']
X = crime[X_columns]
#X_columns
```

```python
# Do something with any missing data.
# Then split your data into test and learning sets.
X.fillna(0, inplace=True)

from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=100,test_size=0.2)
```

```python
# Try using Ordinary Least Squares (sklearn.linear_model.LinearRegression)
# to fit this data.
import sklearn.linear_model
ols = sklearn.linear_model.LinearRegression()
ols.fit(X_train, y_train)
```

```python
# Display the coef_ attribute from your regressor. You might want
# to zip it together with the column names

#print ols.coef_
zip(X_columns, ols.coef_)
```

```python
# How well does it perform as a predictor?
# Hint: try sklearn.metrics.mean_squared_error
print ols.score(X_test,y_test)

ols_predictions = ols.predict(X_test)

from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, ols_predictions)

X_train.columns
```

```python
# Repeat with Lasso, Ridge and ElasticNet
# What happens to the coef_ attributes?
# How much worse are the predictions?

from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error

model_names = ['Lasso', 'Ridge', 'ElasticNet']
models = [Lasso(alpha=0.5), Ridge(alpha=0.5), ElasticNet()]
i = 0

for model in models:
    
    model.fit(X_train, y_train)  
    predictions = model.predict(X_test)
    
    print "Results of {}: ".format(model_names[i])
    print model.score(X_test,y_test)
    print mean_squared_error(y_test, predictions)
#    print model.coef_,'\n'
#    print zip(X_columns, model.coef_)
    print "============================================"
    i +=1
```

```python
# Try using RidgeCV, LassoCV and ElasticNetCV
# to brute-force compromise values.

from sklearn.linear_model import LassoCV
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import ElasticNetCV
from sklearn.metrics import mean_squared_error

model_names = ['LassoCV', 'RidgeCV', 'ElasticNetCV']
models = [LassoCV(), RidgeCV(), ElasticNetCV()]
i = 0

for model in models:
    
    model.fit(X_train, y_train)  
    predictions = model.predict(X_test)
    
    print "Results of {}: ".format(model_names[i])
    print model.score(X_test,y_test)
    print mean_squared_error(y_test, predictions)
#    print model.coef_,'\n'
#    print zip(X_columns, model.coef_)
    print "============================================"
    i +=1
```

```python
# If you scale this data with StandardScaler or RobustScalar, does it make a difference?

import sklearn.preprocessing

scalerX = sklearn.preprocessing.StandardScaler()
scalerX.fit(X)

X_train_scaled = scalerX.transform(X_train)
X_test_scaled = scalerX.transform(X_test)

# optionally transform y
scalerY = sklearn.preprocessing.StandardScaler()
scalerY.fit(y)
Y_train_scaled = scalerY.transform(y_train)
Y_test_scaled = scalerY.transform(y_test)

model_names = ['Lasso', 'LassoCV', 'Ridge', 'RidgeCV', 'ElasticNet', 'ElasticNetCV']
models = [Lasso(alpha=0.5), LassoCV(), Ridge(alpha=0.5), RidgeCV(), ElasticNet(), ElasticNetCV()]
i = 0

for model in models:
    
    model.fit(X_train_scaled, Y_train_scaled)  
    predictions = model.predict(X_test_scaled)
    
    print "Results of {}: ".format(model_names[i])
    print model.score(X_test_scaled,Y_test_scaled)
    print mean_squared_error(Y_test_scaled, predictions)
#    print model.coef_,'\n'
#    print zip(X_columns, model.coef_)
    print "============================================"
    i +=1



#real_predictions = scalerY.inverse_transform(Preds)
```
